# Matplotlib 사용법 
Matplotlib는 파이썬의 대표적인 Visualization 패키지이다. R보다도 다양한 그래프의 표현이 가능하다. 본글에서는 Matplotlib의 기본적인 사용법을 알아보겠다. 
* pyplot 
* figure & subplot 
* 3차원 그래프 그리기 

## pyplot 
기본적 pyplot의 사용법은 아래와 같다. color와 legend는 아래와 같이 설정하고, 마지막에 표시를 하기 위한 별도의 함수를 반드시 써야한다.라벨은 아래와 같이 설정한다. 폰트는 rc함수를 사용한다. 폰트는 한글이 선택되는 "AppleGothic"과 "NamumGothic" 중 하나를 선택하자. 

```python
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt 
matplotlib.rc('font', family="AppleGothic")   
z=[40,50,60,70];y=[10,20,30,40];x=[100,200,300,400]
plt.plot(x,y,color='blue',label='첫번째')
plt.plot(x,z,color='red',label='두번째')
plt.legend()
plt.show()
```

또한 그래프의 lines style을 아래와 같이 조정할 수도 있다. r=red,b=blue,g=green의 색깔을 나타내고, ^은 삼각형, --은 점선을 의미한다. 

```python
t=np.arange(0.,5.,0.2)
plt.plot(t,t,'r^')
plt.plot(t,t**2,'b^')
plt.plot(t,t**3,'g^')
plt.plot(t,t**3.3,'ro--')
plt.show()
```

## figure & subplot 

어떤 그래프에서는 하나의 이미지에 여러가지 그래프를 넣어야 할때가 있다. 있때 우리는 figure를 통해서 상속받은 다음, subplot으로 여러개의 그래프 집합을 완성한다. subplot은 (a,b,c)의 형태로 나뉘는데, a는 가로행, b는 세로열, c는 실제 위치할 위치를 뜻한다. 

```python
import scipy.stats as st 
fig=plt.figure()  
ax1 = fig.add_subplot(3, 1, 1);ax1.hist(st.norm.rvs(size=1000),color='red')
ax2 = fig.add_subplot(3, 1, 2);ax2.hist(st.norm.rvs(size=1000),color='green')
ax3 = fig.add_subplot(3, 1, 3);ax3.hist(st.norm.rvs(size=1000))
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 3, 1);ax1.hist(st.norm.rvs(size=1000),color='green')
ax2 = fig.add_subplot(1, 3, 2);ax2.hist(st.norm.rvs(size=1000),color='red')
ax3 = fig.add_subplot(1, 3, 3);ax3.hist(st.norm.rvs(size=1000))
plt.show()
```

## 3차원 그래프 그리기
3차원 그래프는 Axes3D라는 함수로 표현이 가능하다.Axes3D와 contour함수를 통해서 표현가능하다. 아래의 예는 [헬로데이터 사이언스 스쿨의 결합, 주변, 조건부 확률 밀도 함수](!https://datascienceschool.net/view-notebook/e08074f293144eeca9cc7251b5d5666c/)의 사례이다.

```python
import scipy.stats as st
import numpy as np 

mu=[0,0]
cov=[[2,-1],[-1,18]]
rv=st.multivariate_normal(mu,cov)
xx=np.linspace(-5,5,100)
yy=np.linspace(-5,5,100)
XX,YY=np.meshgrid(xx,yy)
data=np.dstack([XX,YY])
ZZ=rv.pdf(data) 
fig=plt.figure()
ax=Axes3D(fig)
ax.contour(XX, YY, ZZ)
plt.show()
```
